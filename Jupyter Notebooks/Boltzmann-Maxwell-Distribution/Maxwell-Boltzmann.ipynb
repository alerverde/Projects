# Maxwell-Boltzmann Distribution

## Introduction 

The Maxwell-Boltzmann distribution is an important concept in statistical physics and plays a fundamental role in understanding various properties of gases. From this distribution function, the most probable speed, the average speed and the root-mean-square speed can be derived.

According to the Boltzmann equation

$\displaystyle f = Kexp\bigg(-\frac{E}{kT}\bigg)$

where $f$ is the fraction of molecules, $k$ is the Boltzmann constant, $T$ is the thermodynamic temperature, $E$ is the kinetic energy and $K$ is a proportionality constant (at constant temperature). The fraction of molecules with velocity components $v_x,v_y,v_z$ is proportonial to a exponential function of its kinetic energy:

$\displaystyle E = \frac{1}{2}mv_{x}^2 + \frac{1}{2}mv_{y}^2 + \frac{1}{2}mv_{z}^2$

where $m$ is the molecule mass. We can use the relationship $a^{x+y+z+...} = a^x a^y a^z$ so that

$\displaystyle Kexp\bigg(-\frac{\frac{1}{2}mv_{x}^2 + \frac{1}{2}mv_{y}^2 + \frac{1}{2}mv_{z}^2}{kT}\bigg) = Kexp\bigg(-\frac{mv_{x}^2}{2kT}\bigg)exp\bigg(-\frac{mv_{y}^2}{2kT}\bigg)exp\bigg(-\frac{mv_{z}^2}{2kT}\bigg)$

$fv_x,v_y,v_z$ is the fraction of molecules in the velocity interval $v_x$ to $v_x + d_x$, $v_y$ to $v_y + d_y$ and $v_z$ to $v_z + d_z$. We can see how the fraction can be factorized into three factors, one for each axis, and we can write it as $f = f(v_x)f(v_y)f(v_z)$ where

$\displaystyle f(v_x) = K^\frac{1}{3}exp\bigg(-\frac{mv_{x}^2}{2kT}\bigg)$ 

and the same for the other two directions.

In order to determine the constant $K$, considering that the velocity of a molecule is defined in the range $-\infty < v_x < \infty$, we can set up the integral:

$\displaystyle\int\limits_{-\infty}^{\infty} f(v_x)dv_x = 1$

After substituting the expression for $f(v_x)$, we obtain

$\displaystyle 1 = K^\frac{1}{3}\displaystyle\int\limits_{-\infty}^{\infty} exp\bigg(-\frac{mv_{x}^2}{2kT}\bigg) dv_x = K^\frac{1}{3}\bigg(\frac{2\pi kT}{m}\bigg)^\frac{1}{2}$

where we used the standard integral

$\displaystyle\int\limits_{-\infty}^{\infty} exp\bigg(-ax^2\bigg) dx = \frac{\pi}{a}$

Therefore, $ \displaystyle K = (m/2\pi kT)^{3/2}$ and now we can express $f(v_x)$ as:

$\displaystyle f(v_x) = \bigg(\frac{m}{2\pi kT}\bigg)^\frac{1}{2}exp\bigg(\frac{-mv_x^2}{2kT}\bigg)$

The probability for a molecule to have a certain velocity in the intervals $v_x$ to $v_x + d_x$, $v_y$ to $v_y + d_y$ and $v_z$ to $v_z + d_z$ is

$\displaystyle f(v_x)f(v_y)f(v_z)dv_xdv_ydv_z = \bigg(\frac{m}{2\pi kT}\bigg)^\frac{1}{2}exp\bigg(\frac{-mv^2}{2kT}\bigg)dv_xdv_ydv_z$  

where $v^2 = v_x^2+v_y^2+v_z^2$.  

The probability $f(v)dv$ for a molecule to have a velocity between $v$ and $v+dv$, regardless of the direction of motion, is the sum of probabilities for the velocity to be within any of the volume elements $dv_xdv_ydv_z$ that form a spherical shell of radius $v$ and thickness $dv$. The sum of all volume elements on the right side of the equation above represents the volumen of the shell, $4\pi v^2dv$. Thus, we finally have

$ \displaystyle f(v) = 4\pi \bigg[ \frac{m}{2\pi kT}\bigg]^{3/2} v^2 exp\bigg(-\frac{mv^2}{2kT}\bigg) $

This equation has some particular characteristics:

1) The equation includes a decreasing exponential factor, $exp(-\frac{mv^2}{2kT})$ . Its presence implies that the fraction of molecules with high velocities is very small because $exp(-x^2)$ becomes very small when $x^2$ is large.

2) The factor $\frac{m}{(2kT)}$ that multiplies $v^2$ in the exponent becomes significant when the mass of the molecule, m, is large. As a result,  the exponential factor approaches zero rapidly for heavy molecules. This means that heavy molecules with high velocities are rarely found.

3) The opposite happens when the temperature, T, is high: the factor $\frac{m}{2kT}$ in the exponent becomes small, and the exponential factor  approaches zero relatively slowly as $v$ increases. This suggests that we can expect a higher fraction of molecules with high velocities at higher temperatures than at lower temperatures.

4) The $v^2$ factor that multiplies the exponential tends to zero if $v$ tends to zero as well, so the fraction of very slow molecules is also very small.

5) The remaining factors ensure that the sum of fractions from all velocity intervals, ranging from zero to infite, adds up to $1$. 

## Maxwell-Boltzmann Distribution Dependence with Temperature

In order to have a closer look at the Maxwell-Boltzmann distribution dependence with temperature, we can make a graph for Argon at different temperatures.

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
# Constants
m =  0.040 / 6.022e23 #  Ar mass in kg / mol divided by the Avogadro's number
T_values = [25, 100, 300, 1200]  # List of Temperature values in Kelvin
kB = 1.3806503e-23  # m2 kg s-2 K-1 (the Boltzmann constant)
v = np.arange(0.0, 2000, 0.01)  # Set graph limits from 0 to 2000 m/s

# Maxwell-Boltzmann distribution function
def maxwell_boltzmann_distribution(m, T, v):
    return (4. * np.pi * (m / (2. * np.pi * kB * T)) ** 1.5) * v**2 * np.exp((-m * v**2) / (2. * kB * T))

# Plot the figure
plt.figure(figsize=(7, 7))

# Iterate through different temperature values and plot Maxwell-Boltzmann distributions
for T in T_values:
    label = f"{T} K"
    plt.plot(v, maxwell_boltzmann_distribution(m, T, v), label=label)
    plt.fill_between(v, maxwell_boltzmann_distribution(m, T, v), alpha=0.1)

plt.xlabel('v (m/s)', fontsize=14)
plt.ylabel('f(v)', fontsize=16)
plt.title('Maxwell-Boltzmann velocity distribution for Ar')
plt.legend()
plt.show()
```

## Maxwell-Boltzmann Distribution Dependence with Mass

As we did with the temperature relationship, we can also create a graph for different atoms at the same temperature.

```python
# Constants
names = ['Ne', 'Ar','Xenon']
m_elements = [0.020, 0.040, 0.132]  # Atomic masses kg/mol 
atomic_mass_conversion = 6.022e23  # The Avogadro's number
m_elements = [m / atomic_mass_conversion for m in m_elements] # Conversion
T = 300  # Temperature in Kelvin
v = np.arange(0.0, 2000, 0.01)  # Set graph limits from 0 to 2000 m/s

# Plot the figure
plt.figure(figsize=(7, 7))

# Iterate through different mass values and plot Maxwell-Boltzmann distributions
for name, m in zip(names, m_elements):
    label = f"{name}"
    plt.plot(v, maxwell_boltzmann_distribution(m, T, v), label=label)
    plt.fill_between(v, maxwell_boltzmann_distribution(m, T, v), alpha=0.1)

plt.xlabel('v (m/s)', fontsize=14)
plt.ylabel('f(v)', fontsize=16)
plt.title('Maxwell-Boltzmann velocity for diferent atoms')
plt.legend()
plt.show()
```

## Velocity Metrics Derived from the Maxwell-Boltzmann Distribution

The mean velocity $\bar c$, most probable velocity (mode)  $c^*$, and root-mean-square velocity $\sqrt{c^*}$ can be obtained from properties of the Maxwell-Boltzmann distribution. 


The most probable velocity, $\bar c$, is the velocity most likely to be possessed by any molecule (of the same mass m) in the system and corresponds to the maximum value or the mode of $f(v)$. To find it, we calculate the derivative $\frac{df}{dv}$, set it to zero and solve for $v$:

$\displaystyle {\frac{df(v)}{dv}} = -8\pi \bigg[\frac{m}{2\pi kT}\bigg]^{3/2} v \bigg[\frac{mv^2}{2kT}-1\bigg] exp\bigg(-\frac{mv^2}{2kT}\bigg)=0$


with the solution

$\displaystyle \frac{m {c^*}^2}{2kT} = 1$

we finally obtain

$\displaystyle c^* = \sqrt {\frac {2kT}{m}}$

The mean velocity is the expected value of the velocity distribution, setting $b = {\frac{m}{2kT}}$:

$\displaystyle \bar c = \int\limits_{0}^{\infty} vf(v)dv = 4\pi \bigg[\frac{b}{\pi}\bigg]^{3/2} \displaystyle\int\limits_{0}^{\infty}v^3exp\bigg(-bv^2\bigg)dv$

we arrive at

$\displaystyle 4\pi \bigg[\frac{b}{\pi}\bigg]^{3/2} \frac{1}{2b^2}=\sqrt{\frac {4}{\pi b}}$

after replace we obtain

$\displaystyle \bar c = \sqrt{\bigg(\frac{8kT}{\pi m}\bigg)}$

The mean square velocity $\sqrt{\bar c}$ is the second-order raw moment of the velocity distribution. The "root mean square velocity" is the square root of the mean square velocity, corresponding to the speed of a particle with average kinetic energy, setting $b = {\frac{m}{2kT}}$:

$\displaystyle \sqrt{\bar c}=\bigg[\displaystyle\int\limits_{0}^{\infty} v^2 f(v)dv\bigg]^\frac{1}{2}=\bigg[ \displaystyle 4\pi \bigg(\frac{b}{\pi}\bigg)^\frac{3}{2}\displaystyle\int\limits_{0}^{\infty} v^4 exp\bigg(-bv^2\bigg)dv\bigg]^\frac{1}{2}$

we arrive at

$\displaystyle \bigg[ 4\pi \bigg(\frac{b}{\pi}\bigg)^\frac{3}{2}\frac{3}{8} \bigg(\frac{\pi}{b^5}\bigg)^\frac{1}{2}\bigg]^\frac{1}{2} = \sqrt{\frac{3}{2b}}$

we finally obtain

$\displaystyle \sqrt{\bar c} = \bigg(\frac{3kT}{m}\bigg)^\frac{1}{2}$

In the following graph we represent velocity metrics derived from the Maxwell-Boltzmann distribution.

```python
# Plot the figure
plt.figure(figsize=(7, 7))
v = np.arange(0.0, 5, 0.01)  # Set graph limits from 0 to 1

def maxwell_boltzmann_distribution(v):
    return (v**2 * np.exp(-v**2))

plt.plot(v, maxwell_boltzmann_distribution(v))
plt.fill_between(v, maxwell_boltzmann_distribution(v), alpha=0.1)

names = [r'$c^*$', r'$\bar c$', r'$\sqrt{\bar c}$']
values = np.array([1,(4/np.pi)**(1/2),(3/2)**(1/2)])
colors = ['black','red','green']

# Iterate through lists of names, values, and colors to create scatter plots
for name, values, colors in zip(names, values, colors):
    label = f"{name}"
    value = maxwell_boltzmann_distribution(values)
    plt.scatter(y=value,x=values, color=colors, linestyle='-',label=label)

plt.xlabel(r'$v / (2kT/m)^{(1/2)}$', fontsize=14)
plt.ylabel(r'$f(v) / 4\pi (m/2\pi kT)^{(1/2)}$', fontsize=16)
plt.title('Maxwell-Boltzmann velocity distribution')
plt.legend()
plt.show()
```
