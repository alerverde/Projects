The Maxwell-Boltzmann distribution equation forms the basis of the kinetic theory and defines the distribution of speeds for a gas at a certain temperature. 

From this distribution function, the most probable speed, the average speed and the root-mean-square speed can be derived.

According to the equation the fraction of molecules with velocity components $v_x,v_y,v_z$ is proportonial to a exponential funtion of its kinetic energy:

$E = \frac{1}{2}mv_{x}^2 + \frac{1}{2}mv_{y}^2 + \frac{1}{2}mv_{z}^2$

We can use the relationship $a^{x+y+z+...} = a^x a^y a^z$ so that

$f = Kexp\bigg(-\frac{E}{kT}\bigg) = Kexp\bigg(-\frac{\frac{1}{2}mv_{x}^2 + \frac{1}{2}mv_{y}^2 + \frac{1}{2}mv_{z}^2}{kT}\bigg) = Kexp\bigg(-\frac{mv_{x}^2}{2kT}\bigg)exp\bigg(-\frac{mv_{y}^2}{2kT}\bigg)exp\bigg(-\frac{mv_{z}^2}{2kT}\bigg)$

where K is a proportionality constant (at constant temperature) and $fv_{x},v_{y},v_{z}$ is the fraction of molecules in the velocity interval $v_x$ to $v_x + d_x$, $v_y$ to $v_y + d_y$ y $v_z$ to $v_z + d_z$. We can see how the fraction is factorize in three factors, one for each axis, and that we can write it as $f = f(v_x)f(v_y)f(v_z)$ with

$f(v_x) = K^\frac{1}{3}exp\bigg(-\frac{mv_{x}^2}{2kT}\bigg)$ 

and the same for the other two directions.

In order to determinate the K constant, noting that the velocity of a molecule is comprised between $-\infty<v_x<\infty$ so that

$\displaystyle\int\limits_{-\infty}^{\infty} f(v_x)dv_x = 1$

After sustitution of $f(v_x)$ by it's expression, we get

$1 = K^\frac{1}{3}\displaystyle\int\limits_{-\infty}^{\infty} exp\bigg(-\frac{mv_{x}^2}{2kT}\bigg) dv_x = K^\frac{1}{3}\bigg(\frac{2\pi kT}{m}\bigg)^\frac{1}{2}$

where we used the stander integral

$\displaystyle\int\limits_{-\infty}^{\infty} exp\bigg(-ax^2\bigg) dx = \frac{\pi}{a}$

Therefore, $ K = (m/2\pi kT)^{3/2}$ and now

$f(v_x) = \bigg(\frac{m}{2\pi kT}\bigg)^\frac{1}{2}exp\bigg(\frac{-mv_x^2}{2kT}\bigg)$

The probability for a molecule to have a certain velocity in the interval $v_x$ to $v_x + d_x$, $v_y$ to $v_y + d_y$ y $v_z$ to $v_z + d_z$ is

$f(v_x)f(v_y)f(v_z)dv_xdv_ydv_z = \bigg(\frac{m}{2\pi kT}\bigg)^\frac{1}{2}exp\bigg(\frac{-mv^2}{2kT}\bigg)dv_xdv_ydv_z$

where $v^2 = v_x^2+v_y^2+v_z^2$. The probability $f(v)dv$ for a molecule to have a velocity between $v$ and $v+dv$ 

$ f(v) d^3v = \bigg(\frac{m}{2\pi kT}\bigg)^{3/2} exp\bigg(-\frac{mv^2}{2kT}\bigg)d^3v $

where:

*m* is the particle mass,  
*k* is the Boltzmann constant,  
T is the thermodynamic temperature,  
*f(v)* is a probability distribucion function and it's normalize, that means that the $\int f(v)d^3v$ over all velocity is unity.

$$ f(v) = 4\pi v^2 \bigg[ \frac{m}{2\pi kT}\bigg]^{3/2} exp\bigg(-\frac{mv^2}{2kT}\bigg) $$

This equation has some particular characteristics:  
1) The equation includes a decresent exponential, the factor $ exp(-\frac{mv^2}{2kT}) $ .  It's presence implies that the fraction of molecules with high velocities is very small because $exp(-x^2)$ becomes very small when $x^2$ is to large.    
2) The factor $\frac{m}{(2kT)}$ that multiplies $v^2$ in the exponent is too large when the mass of the particle, m, is large; then, the exponential factor gets closer to 0 rapidly when m is large. This means that heavy particles with high velocities are rarely found.  
3) The opposite happens when the temperature, T, is high: the factor $\frac{m}{2kT}$ in the exponent is small, and the exponential factor aproximates to cero in a relative slow way when $v$ increases. This means that we can expect a higher fraction of particules with high velocity at higher temperatures than at lower temperatures.  
4) The $v^2$ factor that multiplies the exponential tends to cero if $v$ tends to cero as well, so the fraction of very slow particles is also very small.  
5) The remaining factors simply ensurence that when the addition of fractions from all the velocity intervals from cero to infite, the result is $1$. 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# A function that calculates the Maxwell-Boltzmann equation
m = 39.948*1.673534e-27
T=300
v = np.arange(0.0, 2000, 0.01) #set graph limits from 0 to 2000 K
kB = 1.3806503e-23
def mbe(m, T, v):
    return 4.*np.pi*v*v*((m/(2.*np.pi*kB*T))**1.5) * np.exp((-m*v*v)/(2.*kB*T))

mbe(m,T,v)
```

```python
# Constants
m =  0.03995 / 6.022e23 #  Ar kg mol divided by the Avogadro's number
T_values = [25, 100, 300, 1200]  # List of Temperature values in Kelvin
kB = 1.3806503e-23  # m2 kg s-2 K-1 (the Boltzmann constant)

# Maxwell-Boltzmann distribution function
def maxwell_boltzmann_distribution(m, T, v):
    return (m / (2. * np.pi * kB * T)) ** 1.5 * 4. * np.pi * v**2 * np.exp((-m * v**2) / (2. * kB * T))

v = np.arange(0.0, 2000, 0.01)  # Set graph limits from 0 to 2000 m/s

# Plot the figure
plt.figure(figsize=(7, 7))

for T in T_values:
    label = f"{T} K"
    plt.plot(v, maxwell_boltzmann_distribution(m, T, v), label=label)
    plt.fill_between(v, maxwell_boltzmann_distribution(m, T, v), alpha=0.1)

plt.xlabel('v (m/s)', fontsize=14)
plt.ylabel('f(v)', fontsize=16)
plt.title('Maxwell-Boltzmann speed distribution for Ar')
plt.legend()
plt.show()
```

```python
# Constants
names = ['Ne', 'Ar','Xenon']
m_elements = [0.020, 0.040, 0.132]  # Atomic masses in atomic mass units (u)
atomic_mass_conversion = 6.022e23  # Conversion factor to kilograms
m_elements = [m / atomic_mass_conversion for m in m_elements]
T = 300  # Temperature in Kelvin

v = np.arange(0.0, 2000, 0.01)  # Set graph limits from 0 to 2000 m/s

# Plot the figure
plt.figure(figsize=(7, 7))

for name, m in zip(names, m_elements):
    label = f"{name}"
    plt.plot(v, maxwell_boltzmann_distribution(m, T, v), label=label)
    plt.fill_between(v, maxwell_boltzmann_distribution(m, T, v), alpha=0.1)

plt.xlabel('v (m/s)', fontsize=14)
plt.ylabel('f(v)', fontsize=16)
plt.title('Maxwell-Boltzmann speed distribution')
plt.legend()
plt.show()
```
